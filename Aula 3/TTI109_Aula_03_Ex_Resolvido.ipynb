## **Exercício 01 - Item (a)**

Use um gráfico de pizza para retratar os resultados de uma pesquisa que investigou as intenções de investimentos de adultos americanos em $2013$, quando comparadas com o ano de $2012$ (Fonte: *CNN*).

* Investir mais em ações: $562$
* Investir mais em títulos: $144$
* Não investir: $288$
* Investir o mesmo valor de $2012$: $461$




```
# Construção do conjunto de dados
info = [['Ações', 562], ['Títulos', 144], ['Não investir', 288],
         ['Igual 2012', 461]]
tabela = pd.DataFrame(info, columns=['Intenções', 'Frequência'])
tabela['Porcentagem'] = tabela['Frequência']/tabela['Frequência'].sum()*100
display(tabela)
```

```
# Gráfico setorial (pizza)
plt.pie(x=tabela['Frequência'], labels=tabela['Intenções'],
                autopct='%1.1f%%', wedgeprops={'edgecolor': 'black'},
                shadow=True)
plt.show()
```

## **Exercício 02 - Item (a)**

Use um gráfico de Pareto para retratar o número de medalhas obtidas por atletas de cinco países nos Jogos Olímpicos de Londres, em $2012$.

* Alemanha: $44$
* Grã-Bretanha: $65$
* Estados Unidos: $104$
* Rússia: $82$
* China: $88$

```
# Construção do conjunto de dados
medalhas = pd.DataFrame({'Medalhas': [44, 65, 104, 82, 88]})
medalhas.index = ['Alemanha', 'Grã-Bretanha', 'Estados Unidos', 'Rússia', 'China']

# Organizando em ordem decrescente
medalhas = medalhas.sort_values(by='Medalhas', ascending=False)
display(medalhas)
```

```
# Construção do gráfico de Pareto
plt.bar(x=medalhas.index, height=medalhas['Medalhas'],
        color='cyan', edgecolor='black')
plt.xlabel('Principais países no quadro de medalhas')
plt.ylabel('Número de medalhas')
plt.title('Medalhas em Londres 2012')
plt.show()
```

## **Exercício 05**

Use um gráfico de série temporal para retratar os dados mostrados nas tabelas a seguir. Os dados representam o número de motocicletas (em milhões) registradas no período de $2000$ a $2011$, nos Estados Unidos. (Fonte: *U.S. Federal Highway Administration*).

image.png

```
# Construção do conjunto de dados
motos = pd.DataFrame({'Ano': list(range(2000, 2012)),
                      'Registros': [4.3, 4.9, 5.0, 5.4, 5.8, 6.2,
                                    6.7, 7.1, 7.8, 7.9, 8.2, 8.3]})
display(motos)
```

```
# Série Temporal
plt.plot(motos['Ano'], motos['Registros'], 'go-')
plt.xlabel('Ano')
plt.xticks(ticks=list(range(2000, 2013, 2)))
plt.ylabel('Número de motociletas [milhões]')
plt.title('Evolução do registro de motociletas nos EUA')
plt.grid()
plt.show()
```
