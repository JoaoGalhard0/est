## **Exercício 01 - Histograma 02**

**(a)** Determine o número de classes adotado para a distribuição de frequência.

*Resposta:* Nota-se que as classes estão representadas pelos respectivos pontos médios. Logo, basta examinar o eixo das abscissas e concluir que existem $7$ classes.

**(b)** Determine a amplitude de classe.

*Resposta:* A amplitude da classe é a distância entre os pontos médios. Assim:

$$ \text{amplitude}_\text{classe}=125-72=178-125=231-178=284-231=337-284=390-337=53\text{ pés}$$

**(c)** Estime a amplitude da distribuição.

*Resposta:* De início, vamos estimar os valores máximo e mínimo da distribuição.

$$v_{mín}=72-\frac{\text{amplitude}_\text{classe}}{2}=72-\frac{53}{2}=45,5\text{ pés}$$

$$v_{máx}=390+\frac{\text{amplitude}_\text{classe}}{2}=390+\frac{53}{2}=416,5\text{ pés}$$

Logo, a amplitude da distribuição é dada por:

$$\text{amplitude}_\text{distribuição}=v_{máx}-v_{mín}=416,5-45,5=371\text{ pés}$$

**(d)** Estime a frequência da classe mais populosa.

*Resposta:* A classe mais populosa é representada pela barra mais alta do histograma. No presente caso, a clase mais populosa é representada pelo ponto médio $125$ pés, com cerca de $23$ ocorrências.

## **Exercício 05 - Conjunto de dados 03**

image.png

**(a)** Construa uma tabela de distribuição de frequência usando o número de classes indicado.

**(b)** Inclua colunas registrando os pontos médios das classes, as frequências relativas e acumuladas.


```
import pandas as pd
# Conjunto de dados
tempos = pd.Series([507, 389, 305, 291, 336, 310, 514, 442,
                    373, 428, 387, 454, 323, 441, 388, 426,
                    411, 382, 320, 450, 309, 416, 359, 388,
                    307, 337, 469, 351, 422, 413])
```

```
# Valores máximo, mínimo e número de classes
t_max = tempos.max()
t_min = tempos.min()
num_classes = 8
print(f'Tempo mínimo da distribuição = {t_min} ms')
print(f'Tempo máximo da distribuição = {t_max} ms')
```

```
from math import ceil
# Tabela de distribuição de frequências
dist = tempos.value_counts(bins=num_classes).sort_index()       # Ordenando por classes
dist = pd.DataFrame(dist)                                       # Transformando em DataFrame
dist = dist.reset_index()                                       # Tornando as classes uma coluna do DataFrame
dist.columns = ['Classe', 'Frequência']                         # Nomeando as colunas
pts_medios = [round(inter.mid, 1) for inter in dist['Classe']]  # Pontos médios, com uma casa decimal
dist['Pontos médios'] = pts_medios                              # Atribuição dos pontos médios à uma coluna da tabela
obs = len(tempos)                                               # Número de observações
dist['Freq. relativa'] = dist['Frequência']/obs                 # Cálculo das frequências relativas
dist['Freq. acumulada'] = dist['Frequência'].cumsum()           # Cálculo das frequências acumuladas
display(dist)
```

**(c)** Construa os histogramas de frequência e de frequência relativa.

```
import matplotlib.pyplot as plt

# Construção do histograma de frequências
histograma = tempos.hist(bins=num_classes, color='yellow',
                         edgecolor='black', grid=False)

# Inserção de atributos ao gráfico
ajuste_eixoy = 2
histograma.set(xlabel='Tempos de reação [ms]',
               ylabel='Frequência - Número de mulheres adultas',
               title='Distribuição de frequências dos tempos de reação',
               xticks=dist['Pontos médios'],
               yticks=range(0, dist['Frequência'].max()+ajuste_eixoy))

# Inserção de rótulos nas barras do histograma - OPCIONAL
barras = histograma.patches
freqs = dist['Frequência']
ajuste_texto = 0.1
for barra, freq in zip(barras, freqs):
    altura = barra.get_height()
    histograma.text(x=barra.get_x()+barra.get_width()/2, y=altura+ajuste_texto,
                    s=freq, ha='center', va='bottom')
plt.show()
```

```
import numpy as np
from matplotlib.ticker import PercentFormatter

# Construção do histograma de frequências relativas (usando matplotlib)
figura = plt.figure()
ax = figura.add_subplot(111)
ax.hist(tempos, bins=num_classes, color='yellow', edgecolor='black',
        weights=np.ones_like(tempos)*100/obs)
ax.yaxis.set_major_formatter(PercentFormatter())
ax.set(xlabel = 'Tempos de reação [ms]',
       ylabel = 'Porcentagem de mulheres adultas',
       title = 'Distribuição de frequências dos tempos de reação',
       xticks = dist['Pontos médios'])

# Inserção de rótulos nas barras do histograma - OPCIONAL
barras = ax.patches
freqs = dist['Freq. relativa']
ajuste_texto = 0.1
for barra, freq in zip(barras, freqs):
    altura = barra.get_height()
    ax.text(x=barra.get_x()+barra.get_width()/2, y=altura+ajuste_texto,
            s=f'{freq*100:.1f}%', ha='center', va='bottom')
plt.show()
```

Analisando o histograma de frequência relativa, é possível responder:
* Classe de menor frequência: Tempo de reação média de $472,2$ ms ($3,3\%$)
* Classe de maior frequência: Tempo de reação média de $416,4$ ms ($20,0\%$)

Qual a proporção dos dados cujos registros:

**(e)** São maiores ou iguais aos associados à classe de maior frequência?

Temos que $(20,0 + 13,3 + 3,3 + 6,7)\% = 43,3\%$ dos tempos de reação são maiores ou iguais a $416,4$ ms.

**(f)** Estão associados à classe de menor frequência?

Temos que $3,3\%$ dos tempos de reação estão associados à classe de tempo médio de reação de $472,2$ ms.

**(d)** Construa o polígono de frequências e a ogiva.


```
## Polígono de frequências - Frequência absoluta
# Cálculo dos pontos médios fictícios
fic_esq = [pts_medios[0]-amp]
print(f'Ponto médio fictício à esquerda = {fic_esq[0]}')
fic_dir = [pts_medios[-1]+amp]
print(f'Ponto médio fictício à direita = {fic_dir[0]}')
```

```
# Construção dos dados com inserção dos pontos médios fictícios
x_data = fic_esq + list(dist['Pontos médios']) + fic_dir
print(f'Valores para eixo x: {x_data}')
y_data = [0] + list(dist['Frequência']) + [0]
print(f'Valores para eixo y: {y_data}')
```

```
# Plotagem do polígono de frequências, com diversos atributos
figura = plt.figure()
ax = figura.add_subplot(111)
ax.plot(x_data, y_data,'go--')
ax.set(xlabel='Tempos de reação [ms]',
       ylabel='Frequência - Número de mulheres adultas',
       title='Distribuição de frequências dos tempos de reação',
       xticks=x_data, yticks=range(0, dist['Frequência'].max()+2, 2))
plt.grid()
```

```
## Polígono de frequências - Frequência relativa
figura = plt.figure()
ax = figura.add_subplot(111)
y_data_rel = [y/sum(y_data)*100 for y in y_data]
ax.plot(x_data, y_data_rel,'go--')
ax.yaxis.set_major_formatter(PercentFormatter())
ax.set(xlabel='Tempos de reação [ms]',
       ylabel='Frequência - Número de mulheres adultas',
       title='Distribuição de frequências dos tempos de reação',
       xticks=x_data)
plt.grid()
```

```
## Ogiva - Frequência absoluta
figura = plt.figure()
ax = figura.add_subplot(111)
ax.plot(dist['Pontos médios'], dist['Freq. acumulada'],'go--')
ax.set(xlabel='Tempos de reação [ms]',
       ylabel='Frequência - Número de mulheres adultas',
       title='Distribuição de frequências dos tempos de reação',
       xticks=dist['Pontos médios'])
plt.grid()
```

```
## Ogiva - Frequência relativa
figura = plt.figure()
ax = figura.add_subplot(111)
ax.plot(dist['Pontos médios'], dist['Freq. acumulada']/obs*100,'go--')
ax.yaxis.set_major_formatter(PercentFormatter())
ax.set(xlabel='Tempos de reação [ms]',
       ylabel='Frequência - Número de mulheres adultas',
       title='Distribuição de frequências dos tempos de reação',
       xticks=dist['Pontos médios'])
plt.grid()
```
